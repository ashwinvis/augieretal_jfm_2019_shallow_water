```python
df_lap = DataFrame(columns=pd_columns)
```

```python
%%capture --no-stdout

for row in df_paths.values:
    key = row[0]
    path = row[1]
    print(path)
    keyparams = keyparams_from_path(path)
    s = pandas_from_path(path, key)
    df_lap = df_lap.append(s, ignore_index=True)
```
